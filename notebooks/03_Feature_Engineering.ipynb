# Feature Engineering

## Objective

The objective of this notebook is to transform the cleaned telecom dataset into a machine learning ready dataset.

Tasks performed:

- Loading cleaned dataset
- Dropping unnecessary columns
- Separating target variable
- One-Hot Encoding
- Label Encoding (if required)
- Feature Scaling using StandardScaler
- Feature Selection
- Saving processed dataset

```python
# =====================================================
# Import Required Libraries
# =====================================================

import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler

print("Libraries imported successfully.")
```

```python
# =====================================================
# Load Cleaned Dataset
# =====================================================

df = pd.read_csv("../data/processed/cleaned_telco_data.csv")

print("Dataset Loaded Successfully")
print()

df.head()
```

```python
# =====================================================
# Dataset Shape and Columns
# =====================================================

print("Dataset Shape :", df.shape)

print("\nTarget Variable")

print(df["Churn Label"].value_counts())
```

## Removing Unnecessary Columns

Some columns contain identifiers or information that should not be used during model training because they do not contribute to prediction or may cause data leakage.

```python
# =====================================================
# Remove Identifier and Target Leakage Columns
# =====================================================

drop_columns = [

    "Customer ID",
    "Country",
    "State",
    "City",
    "Zip Code",
    "Latitude",
    "Longitude",

    "Customer Status",
    "Churn Category",
    "Churn Reason",
    "Churn Score"

]

df = df.drop(columns=drop_columns)

print("Columns Removed Successfully")

print("Dataset Shape :", df.shape)

df.head()
```

## Separating Features and Target Variable

```python
# =====================================================
# Separate Features and Target
# =====================================================

X = df.drop("Churn Label", axis=1)

y = df["Churn Label"]

print("Feature Matrix Shape :", X.shape)
print("Target Shape :", y.shape)
```

## One-Hot Encoding

Categorical variables are converted into numerical format using One-Hot Encoding so that machine learning algorithms can process them.

```python
# =====================================================
# One-Hot Encoding
# =====================================================

categorical_columns = X.select_dtypes(include="object").columns

print("Categorical Columns")

print(categorical_columns)
```

```python
# =====================================================
# Apply One-Hot Encoding
# =====================================================

X = pd.get_dummies(
    X,
    columns=categorical_columns,
    drop_first=True
)

print("One-Hot Encoding Completed")
print()

print("New Feature Shape :", X.shape)

X.head()
```

## Feature Scaling

Machine learning algorithms perform better when numerical features are on a similar scale. StandardScaler standardizes numerical variables by removing the mean and scaling to unit variance.

```python
# =====================================================
# Identify Numerical Columns
# =====================================================

numeric_columns = X.select_dtypes(include=["int64", "float64"]).columns

print("Number of Numerical Features :", len(numeric_columns))

numeric_columns
```

```python
# =====================================================
# Label Encoding Target Variable
# =====================================================

from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

y = label_encoder.fit_transform(y)

print("Label Encoding Completed")

print(np.unique(y))
```

```python
# =====================================================
# Apply StandardScaler
# =====================================================

scaler = StandardScaler()

X[numeric_columns] = scaler.fit_transform(X[numeric_columns])

print("Standard Scaling Completed")

X.head()
```

# Feature Selection

At this stage, features with zero variance are removed because they do not contribute useful information for model training. The remaining features are retained for customer churn prediction.

```python
# =====================================================
# Feature Selection
# Remove Zero Variance Features
# =====================================================

constant_columns = [
    col for col in X.columns
    if X[col].nunique() == 1
]

print("Constant Features Found:")
print(constant_columns)

X = X.drop(columns=constant_columns)

print()
print("Updated Feature Matrix Shape:")
print(X.shape)
```

```python
# =====================================================
# Feature Selection using VarianceThreshold
# =====================================================

from sklearn.feature_selection import VarianceThreshold

selector = VarianceThreshold(threshold=0)

X_selected = selector.fit_transform(X)

selected_features = X.columns[selector.get_support()]

X = pd.DataFrame(X_selected, columns=selected_features)

print("Feature Selection Completed")
print()

print("Selected Features:", X.shape[1])
print("Feature Matrix Shape:", X.shape)
```

```python
# =====================================================
# Final Feature Matrix
# =====================================================

print("Final Feature Matrix Shape")
print(X.shape)

print()

print("Target Distribution")
print(pd.Series(y).value_counts())
```

## Save Processed Dataset

The processed dataset is saved for model training in the next stage of the project.

```python
# =====================================================
# Combine Features and Target
# =====================================================

processed_df = X.copy()

processed_df["Churn Label"] = y

processed_df.head()
```

```python
# =====================================================
# Save Feature Engineered Dataset
# =====================================================

processed_df.to_csv(
    "../data/processed/feature_engineered_data.csv",
    index=False
)

print("Feature Engineered Dataset Saved Successfully")
```

## Summary

Feature engineering has been completed successfully.

The following preprocessing steps were performed:

- Removed unnecessary identifier columns
- Separated features and target variable
- Applied One-Hot Encoding on categorical features
- Standardized numerical features using StandardScaler
- Created the final machine learning dataset
- Saved the processed dataset for model training
